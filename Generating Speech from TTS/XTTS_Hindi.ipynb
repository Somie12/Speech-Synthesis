```
import os
import torch
from TTS.api import TTS

# Check if GPU is available
device = 'cuda' if torch.cuda.is_available() else 'cpu'

# Initialize the TTS model here
tts = TTS(model_name="tts_models/multilingual/multi-dataset/xtts_v2", progress_bar=False).to(device)


# Paths for Google Drive
INPUT_FOLDER = '/content/drive/MyDrive/tts model/selected audio 2/transcript/Transcript hindi'  # Adjust to your folder path
OUTPUT_FOLDER = '/content/audio_outputs'

# Ensure the output folder exists
os.makedirs(OUTPUT_FOLDER, exist_ok=True)

# Iterate through transcript files and generate audio
for filename in os.listdir(INPUT_FOLDER):
    if filename.endswith(".txt"):
        transcript_path = os.path.join(INPUT_FOLDER, filename)

        # Read transcript
        with open(transcript_path, "r", encoding="utf-8") as file:
            text = file.read().strip()

        # Generate audio file path
        audio_filename = os.path.splitext(filename)[0] + ".wav"
        audio_path = os.path.join(OUTPUT_FOLDER, audio_filename)

        # Generate speech
        print(f"Generating audio for: {filename}")
        # Call tts_to_file on the 'tts' instance, not the TTS class
        tts.tts_to_file(text=text, speaker_wav="/content/drive/MyDrive/tts model/01.wav", language="hi", file_path=audio_path)

        print(f"Saved: {audio_path}")
```

```
import shutil
from google.colab import files

# Zip the output folder
shutil.make_archive('/content/audio_outputs', 'zip', OUTPUT_FOLDER)

# Download the zip file
files.download('/content/audio_outputs.zip')
```
