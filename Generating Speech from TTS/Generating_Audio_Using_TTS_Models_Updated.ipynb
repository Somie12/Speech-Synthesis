```python
!pip install pesq librosa numpy

import os
import librosa
import numpy as np
from pesq import pesq

# Paths to your audio in Google Drive
REFERENCE_AUDIO_PATH = "/content/drive/MyDrive/selected language audio/Human Generated Audio/selected mandarin_chinese_audio"
GENERATED_AUDIO_PATH = "/content/drive/MyDrive/selected language audio/Machine Generated Audio/Model 2 - XTTS/Mandarin_Chinese/chinese_audio"

# Function to compute PESQ
def compute_pesq(reference_file, generated_file, sample_rate=16000):
    ref_audio, sr1 = librosa.load(reference_file, sr=sample_rate)
    gen_audio, sr2 = librosa.load(generated_file, sr=sample_rate)

    # Ensure sample rates match
    if sr1 != sample_rate or sr2 != sample_rate:
        ref_audio = librosa.resample(ref_audio, orig_sr=sr1, target_sr=sample_rate)
        gen_audio = librosa.resample(gen_audio, orig_sr=sr2, target_sr=sample_rate)

    # Truncate or pad audio to the same length
    min_length = min(len(ref_audio), len(gen_audio))
    ref_audio = ref_audio[:min_length]
    gen_audio = gen_audio[:min_length]

    # Compute PESQ score
    score = pesq(sample_rate, ref_audio, gen_audio, 'wb')  # 'wb' for wideband
    return score

# Process audio files and compute PESQ
pesq_scores = {}
for file in sorted(os.listdir(REFERENCE_AUDIO_PATH)):  # Ensure correct order
    ref_file = os.path.join(REFERENCE_AUDIO_PATH, file)
    gen_file = os.path.join(GENERATED_AUDIO_PATH, file)

    if os.path.exists(gen_file):
        score = compute_pesq(ref_file, gen_file)
        pesq_scores[file] = score
        print(f"PESQ for {file}: {score:.4f}")
    else:
        print(f"Missing generated audio for {file}")

# Save PESQ scores to a file
pesq_score_file = "/content/drive/MyDrive/selected language audio/Machine Generated Audio/Model 2 - XTTS/Mandarin_Chinese/pesq score chinese_xtts_model.txt"
with open(pesq_score_file, "w") as f:
    for file, score in pesq_scores.items():
        f.write(f"{file}: {score:.4f}\n")
    avg_pesq = np.mean(list(pesq_scores.values()))
    f.write(f"\nAverage PESQ Score: {avg_pesq:.4f}")

print(f"\nAverage PESQ Score: {avg_pesq:.4f}")
print(f"PESQ scores saved to {pesq_score_file}")
```

```python
import torch
from nemo.collections.tts.models import Tacotron2Model, HifiGanModel
import os
import soundfile as sf
import numpy as np

# Make sure we're using GPU if available
device = 'cuda' if torch.cuda.is_available() else 'cpu'
print(f"Using device: {device}")

# Mount Google Drive if not already mounted
from google.colab import drive
drive.mount('/content/drive')

# Load the models
spec_generator = Tacotron2Model.from_pretrained("tts_hi_female_tacotron2")
vocoder = HifiGanModel.from_pretrained(model_name="tts_hifigan")

# Move models to appropriate device
spec_generator = spec_generator.to(device)
vocoder = vocoder.to(device)

# Define input/output directories
INPUT_FOLDER = "/content/drive/MyDrive/selected language audio/transcriptions/hindi_transcriptions"
OUTPUT_FOLDER = "/content/drive/MyDrive/selected language audio/Machine Generated Audio/Model 1 - Tacotron/Hindi/generated_audio_hin_tacotron"

# Ensure the output folder exists
os.makedirs(OUTPUT_FOLDER, exist_ok=True)

# Iterate through transcript files and generate audio
for filename in sorted(os.listdir(INPUT_FOLDER)):
    if filename.endswith(".txt"):
        transcript_path = os.path.join(INPUT_FOLDER, filename)

        # Read transcript
        with open(transcript_path, "r", encoding="utf-8") as file:
            text = file.read().strip()

        # Generate audio file path
        audio_filename = os.path.splitext(filename)[0] + ".wav"
        audio_path = os.path.join(OUTPUT_FOLDER, audio_filename)

        print(f"Generating audio for: {filename}")

        try:
            # Generate spectrogram
            parsed = spec_generator.parse(text)
            spectrogram = spec_generator.generate_spectrogram(tokens=parsed)

            # Convert spectrogram to audio
            audio = vocoder.convert_spectrogram_to_audio(spec=spectrogram)

            # Convert to numpy array and save
            audio_np = audio[0].cpu().detach().numpy()
            sf.write(audio_path, audio_np, samplerate=22050)

            print(f"Saved: {audio_path}")

        except Exception as e:
            print(f"Error processing {filename}: {str(e)}")

print("✅ Audio generation completed!")
```
