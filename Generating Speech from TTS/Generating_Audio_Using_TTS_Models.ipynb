```python
!pip install pesq librosa numpy

import os
import librosa
import numpy as np
from pesq import pesq

# Paths to your audio in Google Drive
REFERENCE_AUDIO_PATH = "/content/drive/MyDrive/selected language audio/Human Generated Audio/selected mandarin_chinese_audio"
GENERATED_AUDIO_PATH = "/content/drive/MyDrive/selected language audio/Machine Generated Audio/Model 2 - XTTS/Mandarin_Chinese/chinese_audio"

# Function to compute PESQ
def compute_pesq(reference_file, generated_file, sample_rate=16000):
    ref_audio, sr1 = librosa.load(reference_file, sr=sample_rate)
    gen_audio, sr2 = librosa.load(generated_file, sr=sample_rate)

    # Ensure sample rates match
    if sr1 != sample_rate or sr2 != sample_rate:
        ref_audio = librosa.resample(ref_audio, orig_sr=sr1, target_sr=sample_rate)
        gen_audio = librosa.resample(gen_audio, orig_sr=sr2, target_sr=sample_rate)

    # Truncate or pad audio to the same length
    min_length = min(len(ref_audio), len(gen_audio))
    ref_audio = ref_audio[:min_length]
    gen_audio = gen_audio[:min_length]

    # Compute PESQ score
    score = pesq(sample_rate, ref_audio, gen_audio, 'wb')  # 'wb' for wideband
    return score

# Process audio files and compute PESQ
pesq_scores = {}
for file in sorted(os.listdir(REFERENCE_AUDIO_PATH)):  # Ensure correct order
    ref_file = os.path.join(REFERENCE_AUDIO_PATH, file)
    gen_file = os.path.join(GENERATED_AUDIO_PATH, file)

    if os.path.exists(gen_file):
        score = compute_pesq(ref_file, gen_file)
        pesq_scores[file] = score
        print(f"PESQ for {file}: {score:.4f}")
    else:
        print(f"Missing generated audio for {file}")

# Save PESQ scores to a file
pesq_score_file = "/content/drive/MyDrive/selected language audio/Machine Generated Audio/Model 2 - XTTS/Mandarin_Chinese/pesq score chinese_xtts_model.txt"
with open(pesq_score_file, "w") as f:
    for file, score in pesq_scores.items():
        f.write(f"{file}: {score:.4f}\n")
    avg_pesq = np.mean(list(pesq_scores.values()))
    f.write(f"\nAverage PESQ Score: {avg_pesq:.4f}")

print(f"\nAverage PESQ Score: {avg_pesq:.4f}")
print(f"PESQ scores saved to {pesq_score_file}")
```

```python
import os

# Path to your folder containing the audio files
folder_path = "/content/drive/MyDrive/TTS LID Audios and PESQ scores /TTS/Human Generated Audio/selected hindi_audio"

# List all files in the directory
files = os.listdir(folder_path)

# Iterate over each file
for file_name in files:
    old_path = os.path.join(folder_path, file_name)

    # Skip directories, only process .wav files
    if not os.path.isfile(old_path) or not file_name.endswith(".wav"):
        continue

    # Check if the file name contains "copy of "
    if "copy of " in file_name:
        new_name = file_name.replace("copy of ", "")  # Remove "copy of "
        new_path = os.path.join(folder_path, new_name)

        # Rename the file
        os.rename(old_path, new_path)
        print(f"Renamed: {file_name} -> {new_name}")
    else:
        print(f"Skipped: {file_name}")
```

```python
import os
import re

# Set folder path
folder_path = "/content/drive/MyDrive/TTS LID Audios and PESQ scores /TTS/Machine Generated Audio/Model 1 - Tacotron/Hindi/generated_audio_in_hin"
# Get all files in the folder
files = os.listdir(folder_path)

# Loop through each file
for file in files:
    old_path = os.path.join(folder_path, file)

    # Fix files like "51.wav.wav" → "51.wav"
    if re.match(r'^\d+\.wav\.wav$', file):
        new_name = file.replace(".wav.wav", ".wav")

    # Fix files like "copy of 101.wav.wav" → "101.wav"
    elif re.match(r'^copy of (\d+)\.wav\.wav$', file, re.IGNORECASE):
        new_name = re.sub(r'^copy of (\d+)\.wav\.wav$', r'\1.wav', file, flags=re.IGNORECASE)

    else:
        continue  # Skip correctly named files

    new_path = os.path.join(folder_path, new_name)

    # Rename only if different
    if old_path != new_path:
        os.rename(old_path, new_path)
        print(f"Renamed: {file} → {new_name}")

print("✅ Renaming complete!")
```

```python
import os
import shutil
import re

# Source and destination folders
src_folder = "/content/drive/MyDrive/TTS LID Audios and PESQ scores /TTS/Machine Generated Audio/Model 1 - Tacotron/Hindi/generated audio(151-500)"
dst_folder = "/content/drive/MyDrive/TTS LID Audios and PESQ scores /TTS/Machine Generated Audio/Model 1 - Tacotron/Hindi/generated_audio_in_hin"

# Ensure destination folder exists
os.makedirs(dst_folder, exist_ok=True)

# Get all files in source folder
files = os.listdir(src_folder)

# Move & rename files
for file in files:
    old_path = os.path.join(src_folder, file)

    # Extract filename numbers (151-500)
    match = re.search(r'(\d+)\.wav', file)
    if match:
        num = int(match.group(1))
        if 151 <= num <= 500:
            # Remove "copy of" if present
            new_name = re.sub(r'^copy\s*of\s*', '', file, flags=re.IGNORECASE).strip()
            new_path = os.path.join(dst_folder, new_name)

            # Move the file
            shutil.move(old_path, new_path)
            print(f"Moved: {file} → {new_name}")

print("✅ Moving and renaming complete!")
```
